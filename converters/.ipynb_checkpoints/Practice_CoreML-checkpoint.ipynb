```python
from PIL import Image

def load_and_scale_image(image_url):
    image = Image.open(urllib.request.urlopen(image_url))
    w,h = image.size
    min_dim = min(w,h)
    x0 = int((w - min_dim)/2)
    y0 = int((h - min_dim)/2)
    box = (x0, y0, x0 + min_dim, y0 + min_dim)
    return image.resize((416,416))

def load_local_and_scale_image(image_path):
    image = Image.open(image_path)
    w,h = image.size
    min_dim = min(w,h)
    x0 = 0 # int((w - min_dim)/2)
    y0 = 0 # int((h - min_dim)/2)
    box = (x0, y0, x0 + min_dim, y0 + min_dim)
    return image.resize((416,416))
```

```python
GRID_SIZE = 13
CELL_SIZE = int(416 / GRID_SIZE)
BOXES_PER_CELL = 5

ANCHORS = [[0.57273, 0.677385], 
           [1.87446, 2.06253], 
           [3.33843, 5.47434], 
           [7.88282, 3.52778], 
           [9.77052, 9.16828]]
```

```python
with open('./models/coco_names.txt', 'r') as f:
    COCO_CLASSES = [c.strip() for c in f.readlines()]
    
COCO_CLASSES[:5]
```

```python
def sigmoid(x):
    k = np.exp(-x)
    return 1 / (1 + k)

def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum()
```

```python
class YoloBox:
    def __init__(self, x, y, w, h):
        self.x = x
        self.y = y
        self.w = w
        self.h = h

class YoloPrediction:
    def __init__(self, yolo_box, label, confidence):
        self.yolo_box = yolo_box
        self.label = label
        self.confidence = confidence
```

```python
#implement decode preds

def decode_preds(raw_preds: []):
    #output shape is (1, 425, 13, 13)
    num_classes = len(COCO_CLASSES)
   
    predictions = []
    for CY in range(GRID_SIZE):
        for CX in range(GRID_SIZE):
            for b in range(BOXES_PER_CELL):
                # each vector has 85 entries for all the 5 anchor boxes thus 5*85 = 425
                # using box shift to get move to get the enteries for next anchor box
                
                box_shift = b * (5 + num_classes)
                
                tx = raw_preds[0,box_shift, CY, CX]
                ty = raw_preds[0,box_shift + 1, CY, CX]
                tw = raw_preds[0,box_shift + 2, CY, CX]
                th = raw_preds[0,box_shift + 3, CY, CX]
                to = raw_preds[0, box_shift + 4, CY, CX]
                
                # calculate 
                # x,y (center of the bounding box) we'll have to take sigmoid of tx
                x = (sigmoid(tx) + CX) * CELL_SIZE
                y = (sigmoid(ty) + CY) * CELL_SIZE
                # w,h (width and height of bounding box)
                w = (np.exp(tw) * ANCHORS[b][0]) * CELL_SIZE
                h = (np.exp(th) * ANCHORS[b][1]) * CELL_SIZE
                
                yolo_box = YoloBox(x, y, w, h)
                # objectness wether object is present in the grid or not
                objectness = sigmoid(to)
                
                # then we'll calculate confidence
                box_confidences = raw_preds[0, (box_shift + 5): (box_shift + 5 + num_classes), CY, CX]
                box_confidences = softmax(box_confidences)
                
                class_id = np.argmax(box_confidences)
                class_label = COCO_CLASSES[class_id]
                
                confidence = box_confidences[class_id] * objectness
                
                #append to predctions
                predictions.append((class_id, confidence, x, y, w, h))
                
    
    return sorted(predictions, key= lambda p:p[1], reverse=True)
                
    
```

```python
import copy
from PIL import ImageDraw, ImageColor

def annotate_image(image, preds, min_score=0.5, top=20):
    annotated_image = copy.deepcopy(image)
    draw = ImageDraw.Draw(annotated_image)
    w,h = image.size
    
    colors = ['red', 'orange', 'yellow', 'green', 'blue', 'white']
    
    for class_id, score, xc, yc, w, h in preds[:top]:
        if score < min_score:
            continue
        class_id = int(class_id)
        label = COCO_CLASSES[class_id]
            
        x0 = xc - (w / 2)
        y0 = yc - (h / 2)
        color = ImageColor.colormap[colors[class_id % len(colors)]]
        draw.rectangle([(x0, y0), (x0 + w, y0 + h)], width=2, outline=color)
        draw.text((x0 + 5, y0 + 5), "{} {:0.2f}".format(label, score), fill=color)
    
    return annotated_image
```

# implement vectorized solution

```python
def softmax_2d(x, axis=1):
    x_max = np.max(x, axis=axis)[:, np.newaxis]
    e_x = np.exp(x - x_max)
    x_sum = np.sum(e_x, axis=axis)[:, np.newaxis]    
    return e_x / x_sum

# def softmax(x):
#     e_x = np.exp(x - np.max(x))
#     return e_x / e_x.sum()
```

```python
# impement a vectorized softmax and sigmoid
def softmax(x, axis=1):
    #get max from each row
    x_max = np.max(x, axis=axis).reshape(-1, 1)
    diff = x - x_max
    n = np.exp(diff)
    d = np.sum(n, axis=axis).reshape(-1, 1)
    return n/d
    
```

```python
# prepare vectorized constants
num_classes = len(COCO_CLASSES)

CX =  np.tile(np.arange(num_classes), reps=(GRID_SIZE,GRID_SIZE,1))

CY = np.tile(np.arange(num_classes), reps=(GRID_SIZE,GRID_SIZE,1))
CY = CY.transpose()

x_ANCHORS = np.array(ANCHORS)[..., 0].reshape(1, 1, BOXES_PER_CELL)
y_ANCHORS = np.array(ANCHORS)[..., 1].reshape(1, 1, BOXES_PER_CELL)
```

```python
#iimplement vectorized output
def decode_vectorized_preds(preds):
    #output is of shape 1, 425, 13, 13
    #transpose
    preds = preds.transpose(0, 2, 3, 1)
    
    # now the shape is 1, 13, 13, 425
    #reshape
    preds = preds.reshape(1, GRID_SIZE, GRID_SIZE, 5, 85)
    
    # operate on tx 
    tx = preds[..., 0]
    
    x = ((CX * sigmoid(tx)) * CELL_SIZE).reshape(-1)
    
    # opearte on ty
    ty = preds[..., 1]
    y = ((CY * sigmoid(ty)) * CELL_SIZE).reshape(-1)
    
    # operate on tw
    tw = preds[..., 2]
    w = (x_ANCHORS * np.exp(tw) * CELL_SIZE).reshape(-1)
    
    # operate on th
    th = preds[..., 3]
    h = (y_ANCHORS * np.exp(th) * CELL_SIZE).reshape(-1)
    
    # operate on to
    to = preds[..., 4]
    objectness = sigmoid(to).reshape(-1)
    
    # confidence
    confidence = preds[..., 5:].reshape(GRID_SIZE**2 * BOXES_PER_CELL, num_classes)
    confidence = softmax_2d(confidence) 
    
    box_ids = np.argmax(confidence, axis=1)
    confidences = confidence.max(axis = 1)
    combined_confdence = confidences*objectness
    
    decoded_preds = np.stack([
        box_ids,
        combined_confdence,
        x,
        y,
        w,
        h
    ]).transpose()
    
    return sorted(list(decoded_preds), key= lambda p: p[1], reverse= True)
    
    
    
    
```

```python
decoded_vec_preds = decode_vectorized_preds(preds)
```
